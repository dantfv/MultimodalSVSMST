## To check/drop possible images downloaded failures.

```python
df = pd.read_csv('street_crawler/output1.csv')
```

```python
 
a_df=df.drop_duplicates(['filename'])[['city','direction','filename','latitude','longitude','name','pano_id','uf']] #unique filename 
 
a_df

```

```python
IMG_FOLDER = 'images/'
 
lista = []  
for f in a_df['filename']:
    file = IMG_FOLDER +f
    if os.path.isfile(file): 
        lista.append(f)    
    
    
new_df = a_df[~a_df.filename.isin(lista)] #image request failed to download into /images/*.jpg
new_df.to_csv('street_crawler/streets_lat_long_failed_request.csv',mode = 'w', index=False) 

new_df2 = a_df[a_df.filename.isin(lista)] #images correctly downloaded to /images/*.jpg
new_df2.to_csv('street_crawler/streets_lat_long_curated.csv',mode = 'w', index=False) 
```
