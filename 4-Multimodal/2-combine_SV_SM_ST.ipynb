#### Using Google Street View
- Reading geographic information from cluster in Vale do Ribeira (df_VR)
- Reading google street view crawler information (df_geo_sv)
- Relate image name from crawler with census tract code and indicators, based on lat/long
- Reading Data Frame of google street view images interpreted by CNN (df_sv_vector)
- Merge images vectors and indicators information (df_sv)

```python
df_census_tracts = pd.read_csv('brasil-census-tracts.csv')
df_census_tracts
```

```python
gdf = gpd.GeoDataFrame(df_census_tracts, crs= 'epsg:3857', geometry=[loads(pgon) for pgon in df_census_tracts['WKT']])
```

```python
gdf3857 = gdf.to_crs(3857)  #coordinates in plain degrees
```

```python
df_VR = pd.read_csv('VR_clusters.csv', delimiter=';')
```

```python
df_VR
```

```python
df_geo_sv = pd.read_csv('geo_sv.csv')
df_geo_sv
```

```python
gdf_geo_sv = gpd.GeoDataFrame(df_geo_sv, crs= 'epsg:3857', geometry=[loads(pgon) for pgon in df_geo_sv['geometry']])
gdf_geo_sv
```

```python
#df_geo_sv['geometry']
#gdf3857['geometry'].contains()
df_geo_sv_join = gdf_geo_sv.sjoin(gdf3857[['geometry','CD_GEOCODI']], how="inner", predicate='intersects')
df_geo_sv_join
```

```python
df_geo_sv_join['Compare'] = df_geo_sv_join['setor'] == df_geo_sv_join['CD_GEOCODI']
df_geo_sv_join
```

```python
df_geo_sv_join[df_geo_sv_join['Compare']==False]
```

```python
df_geo_sv_VR = df_geo_sv_join.merge(df_VR, left_on='CD_GEOCODI', right_on='Cod_setor', how='left')
df_geo_sv_VR
```

```python
df_geo_sv_VR[df_geo_sv_VR['Cod_setor'].isna()][['city','geometry','CD_GEOCODI']]
```

```python
df_geo_sv_VR['setor']=df_geo_sv_VR['CD_GEOCODI']
```
