# Train model with random split fold

```python
kf = KFold(n_splits=5, shuffle=True)
```

```python
def train_model(model):
    

    # Get current date and time
    current_time = time.localtime()

    # Format the date and time in desired format
    fold_name = "FOLD_SV_SM_SAT_GROUPED_RIDGE_RANDOM_{}{}{}_{}{}_{}".format(str(current_time.tm_mday).zfill(2),
                                      time.strftime("%b", current_time)[:3],
                                      str(current_time.tm_year)[2:],
                                      str(current_time.tm_hour).zfill(2),
                                      str(current_time.tm_min).zfill(2),indicator)

    # Create the fold
    os.mkdir(fold_name)
        
    i = 0
    for train_index, test_index in kf.split(X):
         os.mkdir(os.path.join(fold_name,str(i)))
         print("TRAIN:", train_index, "TEST:", test_index)
         X_train, X_test = X.iloc[train_index], X.iloc[test_index]
         y_train, y_test = y.iloc[train_index], y.iloc[test_index]

         model.fit(X_train, y_train)

         yhat = model.predict(X_test)
         SS_Residual = sum((y_test-yhat)**2)       
         SS_Total = sum((y_test-np.mean(y_test))**2)     
         r_squared = 1 - (float(SS_Residual))/SS_Total
         adjusted_r_squared = 1 - (1-r_squared)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
         print('r_squared= ',r_squared, ' adjusted_r_squared= ',adjusted_r_squared)


         filename = os.path.join(fold_name,str(i),'finalized_model.sav')

         pickle.dump(model, open(filename, 'wb'))
        
         result = model.score(X_test, y_test)
         print(result)
         print (model.score(X_test, y_test), 1 - (1-model.score(X_test, y_test))*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1))
         with open(os.path.join(fold_name,str(i),'score.txt'), 'w') as file:
            file.write(str(result))
         np.savetxt(os.path.join(fold_name,str(i),'pred.csv'), yhat, delimiter=",")
         #yhat.to_csv(f'FOLD/{i}/pred.csv', mode='a', index=None)
         np.savetxt(os.path.join(fold_name,str(i),'real.csv'), y_test, delimiter=",")
         #y_test.to_csv(f'FOLD/{i}/real.csv', mode='a', index=None)
         (df[['setor','image_name','common_filename','income','literacy','population','longevity']].iloc[test_index]).to_csv(os.path.join(fold_name,str(i),'data.csv'), mode='a', index=None)   
         i = i + 1
```

```python
r_alphas = np.logspace(0, 5, 100)
model = linear_model.RidgeCV(alphas=r_alphas, scoring='r2')
```

```python
def train_model_fold(model):
     # Get current date and time
    current_time = time.localtime()

    # Format the date and time in desired format
    fold_name = "FOLD_SV_SM_SAT_RIDGE_EXCLUSIVE_{}{}{}_{}{}_{}".format(str(current_time.tm_mday).zfill(2),
                                      time.strftime("%b", current_time)[:3],
                                      str(current_time.tm_year)[2:],
                                      str(current_time.tm_hour).zfill(2),
                                      str(current_time.tm_min).zfill(2),indicator)

    # Create the fold
    os.mkdir(fold_name)
    #i = 0
    for split, (train_index, test_index) in enumerate(gkf_cv.split(balanced_sample, groups=balanced_sample['fold'])):
    #for train_index, test_index in kf.split(df,groups=df['fold'] ):
         os.mkdir(os.path.join(fold_name,str(split)))
         print("TRAIN:", train_index, "TEST:", test_index)
         X_train, X_test = X.iloc[train_index], X.iloc[test_index]
         y_train, y_test = y.iloc[train_index], y.iloc[test_index]

         model.fit(X_train, y_train)

         yhat = model.predict(X_test)
         SS_Residual = sum((y_test-yhat)**2)       
         SS_Total = sum((y_test-np.mean(y_test))**2)     
         r_squared = 1 - (float(SS_Residual))/SS_Total
         adjusted_r_squared = 1 - (1-r_squared)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
         print('r_squared= ',r_squared, ' adjusted_r_squared= ',adjusted_r_squared)


         filename = os.path.join(fold_name,str(split),'finalized_model.sav')

         pickle.dump(model, open(filename, 'wb'))
        
         result = model.score(X_test, y_test)
         print(result)
         print (model.score(X_test, y_test), 1 - (1-model.score(X_test, y_test))*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1))
         with open(os.path.join(fold_name,str(split),'score.txt'), 'w') as file:
            file.write(str(result))
         np.savetxt(os.path.join(fold_name,str(split),'pred.csv'), yhat, delimiter=",")
         #yhat.to_csv(f'FOLD/{i}/pred.csv', mode='a', index=None)
         np.savetxt(os.path.join(fold_name,str(split),'real.csv'), y_test, delimiter=",")
         (balanced_sample[['setor','image_name','filename','income','literacy','population','longevity','fold']].iloc[test_index]).to_csv(os.path.join(fold_name,str(split),'data.csv'), mode='a', index=None)   
         #i = i + 1
```
